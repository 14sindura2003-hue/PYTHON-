Write a program to check whether a given number is a palindrome or not.
*

```python
num = 212
rev = 0
temp = num
while num>0:
  rem = num % 10
  rev = rev*10 + rem
  num = num//10
print(rev)
if(temp == rev):
  print("palindrome")
else:
  print("not palindrome")
```

Write a program to check whether a given number is an Armstrong number or not.

```python
num = int(input("enter the number"))
temp = num
sum = 0
l = len(str(num))
for i in range(l):
  rem = num % 10
  sum = sum + rem**l
  num = num//10
print(sum)
if(temp == sum):
  print("armstrong")
else:
  print("not armstrong")
```

Write a program to check whether a given number is prime or not.  


```python
num = int(input("Enter the number: "))
c = 0
for i in range(1, num + 1):
    if num % i == 0:
        c += 1
if c == 2:
    print("It is a prime number")
else:
    print("It is not a prime number")
```

Write a program to find the factorial of a given number.


```python
digit = int(input("enter the number"))
fact = 1
for i in range(1, digit + 1):
  fact = fact * i
print(fact)
```

Write a program to print the first N terms of the Fibonacci series.


```python
n = int(input("Enter the number of terms: "))
a = 0
b = 1
for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```

Write a program to reverse a given string s="Programming"


```python
s = "Programming"
rev = ""
for ch in s:
    rev = ch + rev
print(rev)
```

Write a program to count the number of vowels in a given string. string="Hello"

```python
s = "Hello"
count = 0
for ch in s:
    if ch.lower() in "aeiou":
        count += 1
print("Number of vowels:", count)
```

Write a program to find the sum of digits of a given number.
*

```python
num = int(input("Enter the number: "))
sum = 0
while num > 0:
    digit = num % 10
    sum += digit
    num //= 10
print("The sum of digits is:", sum)
```

Write a Python program to print the multiplication table of a given number up to 10.

```python
num = int(input("Enter the number: "))
for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```
