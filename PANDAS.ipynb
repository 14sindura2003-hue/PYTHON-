Pandas is a powerful Python library used for data analysis and data manipulation. It helps organize, clean, analyze, and process data efficiently.

**Why use Pandas?**
1. Store data in a table format.
2. Read data from CSV, Excel, SQL, etc.
3. Clean missing or incorrect data.
4. Filter and sort data.
5. Perform calculations and analysis.
6. Prepare data for Machine Learning.




**Main Data Structures in Pandas**
1. Series

A Series is a one-dimensional array that stores a single column of data.

```python
import pandas as pd

marks = pd.Series([78, 85, 92, 67, 88])

print(marks)
```

2. DataFrame

A DataFrame is a two-dimensional table with rows and columns.

```python
import pandas as pd

data = {
    "Student": ["A", "B", "C"],
    "Marks": [78, 85, 92]
}

df = pd.DataFrame(data)

print(df)
```

```python
import pandas as pd

students = ["A", "B", "C", "D"]
marks = [78, 85, 92, 67]

df = pd.DataFrame({
    "Student": students,
    "Marks": marks
})

print(df)

print("\nHighest Marks:")
print(df["Marks"].max())

print("\nAverage Marks:")
print(df["Marks"].mean())
```

s.map(func)
map() applies a function to each element of the Series and returns a new Series.

s.map(function)

```python
import pandas as pd

s = pd.Series([10, 20, 30, 40])

result = s.map(lambda x: x + 5)

print(result)
```

s.apply(func)
apply() also applies a function to every element in the Series.


```python
import pandas as pd

s = pd.Series([2, 4, 6, 8])

result = s.apply(lambda x: x * 2)

print(result)
```

s.sort_values()
Sorts the values in a Series.
Ascending order by default



```python
import pandas as pd
s = pd.Series([45, 20, 60, 10])
print(s.sort_values())
#Decending order
print(s.sort_values(ascending=False))
```

s.drop(labels)

**Indexing, Slicing, Filtering, and Aggregation in Pandas Series**

```python
import pandas as pd

# Create a Series
s = pd.Series([78, 85, 92, 67, 88])

print("Original Series:")
print(s)

# Indexing
print("\nIndexing")
print("First Element:", s[0])

# Slicing
print("\nSlicing")
print(s[1:4])

# Filtering
print("\nMarks Greater Than 80")
print(s[s > 80])

# Aggregation
print("\nAggregation Functions")
print("Sum:", s.sum())
print("Average:", s.mean())
print("Maximum:", s.max())
print("Minimum:", s.min())
print("Count:", s.count())
```

**DataFrame in Pandas**

A DataFrame is a two-dimensional table in Pandas that consists of rows and columns. It is the most commonly used data structure in Pandas for storing and analyzing data.

Rows represent records.
Columns represent attributes or fields.
Each column can contain different data types (integers, strings, floats, etc.).




```python
import pandas as pd

# Create a DataFrame
data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Marks": [78, 85, 92, 67, 88],
    "Age": [20, 21, 20, 22, 21]
}

df = pd.DataFrame(data)

# Display DataFrame
print("Original DataFrame")
print(df)

# Display first 5 rows
print("\nHead")
print(df.head())

# Display last 5 rows
print("\nTail")
print(df.tail())

# Display shape
print("\nShape")
print(df.shape)

# Display column names
print("\nColumns")
print(df.columns)

# Display information
print("\nInfo")
print(df.info())

# Display statistical summary
print("\nDescribe")
print(df.describe())

# Access a single column
print("\nStudent Column")
print(df["Student"])

# Access multiple columns
print("\nStudent and Marks")
print(df[["Student", "Marks"]])

# Add a new column
df["Grade"] = ["B", "A", "A+", "C", "A"]

print("\nAfter Adding Grade Column")
print(df)

# Rename a column
df.rename(columns={"Marks": "Score"}, inplace=True)

print("\nAfter Renaming Marks to Score")
print(df)

# Filter data
print("\nStudents with Score greater than 80")
print(df[df["Score"] > 80])

# Sort data in ascending order
print("\nSort by Score (Ascending)")
print(df.sort_values(by="Score"))

# Sort data in descending order
print("\nSort by Score (Descending)")
print(df.sort_values(by="Score", ascending=False))

# Aggregation functions
print("\nMaximum Score:", df["Score"].max())
print("Minimum Score:", df["Score"].min())
print("Average Score:", df["Score"].mean())
print("Total Score:", df["Score"].sum())
print("Count:", df["Score"].count())

# Delete Grade column
df.drop("Grade", axis=1, inplace=True)

print("\nAfter Deleting Grade Column")
print(df)
```
