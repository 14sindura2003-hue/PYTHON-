**LIST **
A list is an ordered, mutable collection that allows duplicate values.
Features:
Ordered
Mutable (can be changed)
Allows duplicates

```python
lst = [10, 20, 30]
print(lst)
lst.append(40)
print(lst)
lst.insert(1, 15)
print(lst)
lst.remove(20)
print(lst)
lst.pop()
print(lst)
lst.sort()
print(lst)
lst.reverse()
print(lst)
```

**Tuple**
A tuple is an ordered, immutable collection.
Features:
Ordered
Cannot be modified
Allows duplicates

```python
t = (10, 20, 30, 20)
print(t)
print(t.count(20))
print(t.index(30))
print(len(t))
```

**set**
A set is an unordered collection of unique elements.
Features:
Unordered
No duplicates
Mutable

```python
s = {10, 20, 30}
print(s)
s.add(40)
print(s)
s.remove(20)
print(s)
s.union({40, 50, 60})
print(s)
s.intersection({40, 50, 60})
print(s)
s.difference({40, 50, 60})
print(s)
```

**dictionary**
A dictionary stores data as key-value pairs.
Features:
Stores data in key-value pairs
Mutable
Keys must be unique

```python
d = {"name":"sindura", "age":21}
print(d)
print(d.keys())
print(d.values())
d.update({"city":"Hyderabad"})
print(d)
d.pop("age")
print(d)
#print(d.get("age"))
d.items()
```
