create a dataset containing marks of 10 students  
students = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]
marks = [78,85,92,67,88,95,72,81,90,76]
Tasks :
1. Display student marks gradually
2. Find the highest score   
3. Add a appropriate title and axis labels
4. Display marks in descending oreder and plot again matplotlib








```python
import matplotlib.pyplot as plt

# Dataset
students = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]
marks = [78, 85, 92, 67, 88, 95, 72, 81, 90, 76]

# Task 1: Display student marks gradually
print("Student Marks")
for i in range(len(students)):
    print(students[i], ":", marks[i])

# Task 2: Find the highest score
highest = max(marks)
index = marks.index(highest)

print("\nHighest Score")
print("Student", students[index], "scored", highest, "marks")

# Task 3: Plot the original marks
plt.figure(figsize=(8,5))
plt.bar(students, marks)

plt.title("Marks of 10 Students")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()

# Task 4: Display marks in descending order
sorted_data = sorted(zip(marks, students), reverse=True)

print("\nMarks in Descending Order")
for mark, student in sorted_data:
    print(student, ":", mark)

# Separate the sorted data for plotting
sorted_marks = []
sorted_students = []

for mark, student in sorted_data:
    sorted_marks.append(mark)
    sorted_students.append(student)

# Task 5: Plot the sorted marks
plt.figure(figsize=(8,5))
plt.bar(sorted_students, sorted_marks)

plt.title("Student Marks in Descending Order")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```
