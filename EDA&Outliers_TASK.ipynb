Task:
You are given height_weight.csv file which contains heights and weights of 1000 people. Dataset is taken from here, https://www.kaggle.com/mustafaali96/weight-height

You need to do this,

(1) Load this csv in pandas dataframe and first plot histograms for height and weight parameters

(2) Using IQR detect weight outliers and print them

(3) Using IQR, detect height outliers and print them

(4) replace the outliers with mean and median and print the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/weight-height.csv')
```

```python
print(df.head())
```

```python
print(df.describe())
```

**HISTOGRAM FOR** **HEIGHT**

```python
plt.figure(figsize=(6,4))
plt.hist(df["Height"], bins=20, color="skyblue", edgecolor="black")
plt.title("Histogram of Height")
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.show()
```

**HISTOGRAM FOR WEIGHT**

```python
plt.figure(figsize=(6,4))
plt.hist(df["Weight"], bins=20, color="lightgreen", edgecolor="black")
plt.title("Histogram of Weight")
plt.xlabel("Weight")
plt.ylabel("Frequency")
plt.show()
```

**Using IQR detect weight outliers and print them**

```python
Q1 = df["Weight"].quantile(0.25)
Q3 = df["Weight"].quantile(0.75)
IQR  = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
weight_outliers = df[(df["Weight"] < lower_limit) | (df["Weight"] > upper_limit)]
print("\nWeight Outliers:")
print(weight_outliers)
```

**Using IQR, detect height outliers and print them**

```python
Q1 = df["Weight"].quantile(0.25)
Q3 = df["Weight"].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
weight_outliers = df[(df["Weight"] < lower_limit) | (df["Weight"] > upper_limit)]
print("\nWeight Outliers:")
print(weight_outliers)
```

**replace the outliers with mean**

```python
df_mean = df.copy()
weight_mean = df_mean["Weight"].mean()
height_mean = df_mean["Height"].mean()
df_mean.loc[(df_mean["Weight"] < lower_limit) | (df_mean["Weight"] > upper_limit), "Weight"] = weight_mean
df_mean.loc[(df_mean["Height"] < lower_limit) | (df_mean["Height"] > upper_limit), "Height"] = height_mean
print("\nDataset after replacing outliers with Mean:")
print(df_mean)
```

**replace the outliers with median**

```python
df_median = df.copy()
weight_median = df_median["Weight"].median()
height_median = df_median["Height"].median()
df_median.loc[(df_median["Weight"] < lower_limit) | (df_median["Weight"] > upper_limit), "Weight"] = weight_median
df_median.loc[(df_median["Height"] < lower_limit) | (df_median["Height"] > upper_limit), "Height"] = height_median
print("\nDataset after replacing outliers with Median:")
print(df_median)
```
